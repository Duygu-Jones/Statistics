# **Correlation & Covariance**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# SciPy

```python
Ekran_sure = [3,5,2,0.5,5,3,1,4,3,4]
AGNO = [2.7,2.2,3.3,3.4,2,3,3.6,2.4,3.3,2.6]
```

### Pearson Formule:



```python
# ilk deger r degeri , 2. deger p-value
# pearson r degerini bulduk =-0.92

stats.pearsonr(Ekran_sure, AGNO)
```

```python
stats.pearsonr(AGNO,Ekran_sure)

# x ve y degerlerinin yerini degistirince degerler yine ayni corr verir (linear da ise positional dikkat)
```

```python
# spearman correaaltion coefficient
stats.spearmanr(Ekran_sure,AGNO)

# data seti normal daglma uygun degilse spearman kullanlabiir
```

# Numpy

```python
# coveriance degerini bulalum

np.cov(AGNO, Ekran_sure)

#cov. degeri -0.78 dir
```

```python
# correlation Coeffition

np.corrcoef(AGNO, Ekran_sure) # kosegenden gecen degerler zaten kendileri ile olan iliskidir

# r = -o.92 bulunur
```

# Linear Regression

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html 

```python
# Linear Regression denklemi icin x vve y nin siralamais onemlidir
# Scipy lingres modulu icin ilgili doc da yazilidir

stats.linregress(Ekran_sure, AGNO)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(Ekran_sure, AGNO)

# yukarda fonksiyon ile return edilen degerleri variable lara atadk
```

```python
# imdi burada bir prediction yapalm..
# x_p ksm hesaplamak istediimiz x deerleridir
# y_p de denklem sonucu hesaplanack sonular

x_p = np.array([0,1,2,3,4,5])
y_p = intercept + slope * x_p
```

```python
#linspace ile 0-5 arasnda eit aralkl 100 deer olsun
# Bu, genellikle bir izgi grafii izmek iin dzgn bir x eksen deerleri seti oluturmak amacyla yaplr.
# slope ve intercept degerleri yukardan alnyor

x_new = np.linspace(0,5,100)
y_new = intercept + slope * x_new
```

```python
#burada data setini izdirip zerine de st satrdaki line  fitted line olarak  izdireceiz

plt.figure(figsize=(8,6))
ax = plt.axes()     #
ax.scatter(Ekran_sure, AGNO)
ax.plot(x_new, y_new)
ax.set_xlabel("Ekran Sure")
ax.set_ylabel("AGNO")
ax.axis("tight")
plt.show()
```
